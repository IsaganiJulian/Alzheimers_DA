```python
#Split rows
def split_values(row):
    if row['Data_Value_Type'] == 'Percentage':
        return row['Data_Value']
    elif row['Data_Value_Type'] == 'Mean':
        return None 
    
# Create Percentage and Mean Columns
df['Percentage'] = df.apply(split_values, axis=1)
df['Mean'] = df.apply(lambda row: row['Data_Value'] if row['Data_Value_Type'] == 'Mean' else None, axis=1)

#Drop Data_Value_Type and Data_Value columns
df = df.drop(['Data_Value_Type', 'Data_Value'], axis=1)

# Save 
alz_df.to_csv('new_dataset.csv', index=False)


```

```python
# Create mapping dictionary 
mapping_dict = {
    'AGE_OVERALL': '',
    '65PLUS' : '65+',
    '5064' : '50-64'
}

#Replace values
df['StratificationID1'] = df['StratificationID1'].replace(mapping_dict)

#Drop StratificationCategoryID1 and StratificationCategory1
df = df.drop(['StratificationCategoryID1', 'StratificationCategory1'], axis=1)

print(df.head(10))

#Save new df to a new csv
df.to_csv('new_dataset.csv', index=False)
```

```python
# Check if 'StratificationCategoryID2' exists in columns
if 'StratificationCategoryID2' in df.columns:
    # Split 'StratificationCategoryID2' into 'Race' and 'Gender'
    split_values_category = df['StratificationCategoryID2'].str.split('_', expand=True)

    # Ensure that split_values_category has two columns
    if split_values_category.shape[1] == 2:
        df[['Race', 'Gender']] = split_values_category
    else:
        # Handle the case where split_values_category has a different number of columns
        print("Error: Unexpected number of split values in StratificationCategoryID2")

    # Drop the original 'StratificationCategoryID2' column if needed
    df = df.drop(['StratificationCategoryID2'], axis=1)
else:
    print("Error: 'StratificationCategoryID2' column not found in the DataFrame.")

# Check if 'StratificationID2' exists in columns
if 'StratificationID2' in df.columns:
    # Split 'StratificationID2' into 'Race' and 'Gender'
    split_values_id = df['StratificationID2'].str.split('_', expand=True)

    # Ensure that split_values_id has two columns
    if split_values_id.shape[1] == 2:
        df[['Race', 'Gender']] = split_values_id
    else:
        # Handle the case where split_values_id has a different number of columns
        print("Error: Unexpected number of split values in StratificationID2")

    # Drop the original 'StratificationID2' column if needed
    df = df.drop(['StratificationID2'], axis=1)
else:
    print("Error: 'StratificationID2' column not found in the DataFrame.")

# Display the updated dataframe
print(df.head(10))

# Save the new dataframe to a new CSV file
df.to_csv('new_dataset.csv', index=False)
```
