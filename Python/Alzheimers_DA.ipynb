```python
#Split rows
def split_values(row):
    if row['Data_Value_Type'] == 'Percentage':
        return row['Data_Value']
    elif row['Data_Value_Type'] == 'Mean':
        return None 
    
# Create Percentage and Mean Columns
df['Percentage'] = df.apply(split_values, axis=1)
df['Mean'] = df.apply(lambda row: row['Data_Value'] if row['Data_Value_Type'] == 'Mean' else None, axis=1)

#Drop Data_Value_Type and Data_Value columns
df = df.drop(['Data_Value_Type', 'Data_Value'], axis=1)

df
```

```python
#fillna for percentage and mean columns
df = df.fillna('')
df.to_csv('new_dataset.csv', index=False)
```

```python
# Drop columns
columns_to_drop = [
    'RowId', 'LocationAbbr', 'Response',
    'Data_Value_Unit', 'Data_Value_Alt', 'Data_Value_Footnote_Symbol', 
    'Data_Value_Footnote', 'Sample_Size', 'StratificationCategory3',
    'Stratification3', 'ResponseID', 'StratificationCategoryID3',
    'StratificationID3', 'Report', 'DataValueTypeID', 'StratificationID1', 
    'StratificationCategoryID1', 'StratificationCategory1'
]

# To nsure that the columns to drop actually exist in the DataFrame
columns_to_drop = [col for col in columns_to_drop if col in df.columns]

# Drop columns
df = df.drop(columns=columns_to_drop, inplace=False)

# Display the updated dataframe
df

# Save
df.to_csv('new_dataset.csv', index=False)

```

```python


# 1. Handling Missing Values
df['LocationDesc'].fillna('Unknown', inplace=True)
df['Age_Group'].fillna('Unknown', inplace=True)

# 2. Convert Age_Group to Categorical
df['Age_Group'] = pd.Categorical(df['Age_Group'])

# 3. Encoding Boolean Values
bool_columns = ['Race/Ethnicity_Asian/Pacific Islander', 'Race/Ethnicity_Black, non-Hispanic',
                'Race/Ethnicity_Hispanic', 'Race/Ethnicity_Native Am/Alaskan Native',
                'Race/Ethnicity_White, non-Hispanic', 'Gender_Female', 'Gender_Male']

df[bool_columns] = df[bool_columns].astype(bool)

# 4. Drop unnecessary columns
columns_to_drop = ['YearEnd', 'Datasource', 'Class', 'Topic']
df.drop(columns=columns_to_drop, inplace=True)

# 5. Rename columns 
column_mapping = {
    'Race/Ethnicity_Asian/Pacific Islander': 'Asian/Pacific Islander',
    'Race/Ethnicity_Black, non-Hispanic': 'Black, non-Hispanic',
    'Race/Ethnicity_Hispanic': 'Hispanic',
    'Race/Ethnicity_Native Am/Alaskan Native': 'Native Am/Alaskan Native',
    'Race/Ethnicity_White, non-Hispanic': 'White, non-Hispanic',
    'Gender_Female': 'Female',
    'Gender_Male': 'Male'
}

df.rename(columns=column_mapping, inplace=True)

# Display the cleaned and formatted data
print("\nCleaned and Formatted Data:")
print(df.head())
# Save the data to a new CSV file
df.to_csv('df.csv', index=False)
```
