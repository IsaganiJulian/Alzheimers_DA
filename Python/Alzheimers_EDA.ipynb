
# Question Group: Arthritis Distinct Questions:
  -Percentage of older adults ever told they have arthritis

  -Fair or poor health among older adults with doctor-diagnosed arthritis
  
  -Severe joint pain due to arthritis among older adults with doctor-diagnosed arthritis

```python
# Filter the DataFrame for questions grouped as Arthritis and the specified age groups
arthritis_df = df[
    (df['Question_Group'].str.lower() == 'arthritis') & 
    (df['Age_Group'].isin(['50-64 years', '65 years or older'])) & 
    (df['Percentage'].notna())
    ]

# Convert 'Percentage' column to numeric if it's not already
arthritis_df['Percentage'] = pd.to_numeric(arthritis_df['Percentage'], errors='coerce')

# Group by 'Age_Group' and 'Question', and calculate the average percentage
average_percentage_by_age_question = arthritis_df.groupby(['Age_Group', 'Question'])['Percentage'].mean().reset_index()

# Iterate over unique questions and create plots
for question in average_percentage_by_age_question['Question'].unique():
    # Filter data for the specific question
    question_data = average_percentage_by_age_question[average_percentage_by_age_question['Question'] == question]
    # Plot
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Age_Group', y='Percentage', data=question_data, palette='viridis')
    plt.title(f'Average Percentage for {question} by Age Group')
    plt.xlabel('Age Group')
    plt.ylabel('Average Percentage')
    plt.show()
```
